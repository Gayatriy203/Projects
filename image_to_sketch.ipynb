```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow 
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt 
```

```python
from google.colab.patches import cv2_imshow
image = io.imread("/content/drive/MyDrive/data/amir-esrafili-X8Cao0DbiDg-unsplash.jpg")
cv2_imshow(image)
```

```python
%matplotlib inline
img_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
gray_image = cv.cvtColor(image , cv.COLOR_BGR2GRAY)
gray_image
```

```python
plt.imshow(gray_image)
```

inverting image

```python
inverted_image = 255 - gray_image
plt.imshow(inverted_image)
```

```python
blurred = cv.GaussianBlur(inverted_image,(21,21),0)
plt.imshow(blurred)
```

sketch image

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv.divide(gray_image,inverted_blurred,scale=255.0)
plt.imshow(pencil_sketch)
cv.imwrite("sketch.jpg",inverted_blurred)
```
